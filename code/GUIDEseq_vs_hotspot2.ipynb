```python
## define function: calculate distance to the closest Hotspot for a given GUIDE-seq hit position

def disMinHotspotCal(dfGUI, dfHOT):
    disMinNeighborHotspot =[]
    for row in dfGUI.iterrows():
        currentChr = row[1]['#BED Chromosome']
        position = row[1]['Position']
        disList =list(position - dfHOT['midPos'][dfHOT['chr'] == currentChr])
        disListABS = [abs(number) for number in disList]
        #print(disListABS.shape)
        disMin = min(disListABS)

        #inde = disListABS[disListABS == disMin].index[0]
        #inde.reset_index(drop=True, inplace=True)
        inde = disListABS.index(disMin)

        #print(position, currentChr, len(disList), disMin, inde, disList[inde], disListABS[inde])
        halfWidth = (dfHOT['Width'][dfHOT['chr'] == currentChr].iloc[inde]) /2
        matchedIndex = dfHOT[dfHOT['chr'] == currentChr].iloc[inde]
        #print(halfWidth, dfHOTSPOT[dfHOTSPOT['chr'] == currentChr].iloc[inde])

        if disListABS[inde] < halfWidth:
            disMinNeighborHotspot.append(0)
        else:
            distance= disListABS[inde] - halfWidth
            disMinNeighborHotspot.append(distance)
    #print disMinNeighborHotspot
    return(disMinNeighborHotspot)
```

```python
### Special variable for linear regression
### Thought: gRNA binding frequency is negatively associated with hotspot proximity
### Thought2: middle of hotspot has the strongest effect on gRNA binding frequency

def disMinHotspotMiddle(dfGUI, dfHOT):
    
    disMinNeighborHotspot =[]
    for row in dfGUI.iterrows():
        currentChr = row[1]['#BED Chromosome']
        position = row[1]['Position']
        disList =list(position - dfHOT['midPos'][dfHOT['chr'] == currentChr])
        disListABS = [abs(number) for number in disList]
        #print(disListABS.shape)
        disMin = min(disListABS)

        #inde = disListABS[disListABS == disMin].index[0]
        #inde.reset_index(drop=True, inplace=True)
        inde = disListABS.index(disMin)

        #print(position, currentChr, len(disList), disMin, inde, disList[inde], disListABS[inde])
        halfWidth = (dfHOT['Width'][dfHOT['chr'] == currentChr].iloc[inde]) /2
        matchedIndex = dfHOT[dfHOT['chr'] == currentChr].iloc[inde]
        #print(halfWidth, dfHOTSPOT[dfHOTSPOT['chr'] == currentChr].iloc[inde])

        distance = disListABS[inde]
        disMinNeighborHotspot.append(distance)
    #print disMinNeighborHotspot
    return(disMinNeighborHotspot)
```
